```python
#given
h_o = 1.2
v_o = 5.4
g = 9.8

def height(t):
  h = h_o + v_o * t - 0.5 * g * t * t
  return h

def velocity(t):
  v = v_o - g * t
  return v
```

```python
def fact_while(n):
  fact = 1
  i = 1
  while(i <= n):
    fact = fact * i
    i = i + 1
  return fact

def fact_for(n):
  fact = 1
  for i in range(1 , n+1):
    fact = fact * i
  return fact

data = [[i , fact_while(i) , fact_for(i)] for i in range(0,6)]
print(tabulate(data , headers=['Num' , 'fact_while' , 'fact_for']))
```
