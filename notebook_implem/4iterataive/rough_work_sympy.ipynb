```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
x = sy.Symbol('x')
```

```python
f = (2*x**3+1 ) / (3*x**2 - 1)
```

```python
f
```

```python
f_diff = sy.diff(f,x)
```

```python
sy.simplify(f_diff)
```

```python
sy.plot(f_diff , (x , 0 , 2), title='phi_derivative' , ylim = (-100 , 100) )
```

# for some reason the root of derivative of phi(x) is the root of f(x) f the phi(x) is determined by the newton rhapson method. HOW????


```python
f_diff.subs(x,1.1)
```
