```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
from tabulate import tabulate
%matplotlib notebook
```

```python
def fixed_itertation(g , x_0 , eps , max_steps=30 , verbose=True):
    for _ in range(0,max_steps):
        x_1 = g(x_0)
        if verbose==True:
            print(x_1)
        if(abs(x_1 - x_0) < eps):
            return x_1
        x_0 = x_1
    
    print('max steps exceeded')
    return np.nan
```
