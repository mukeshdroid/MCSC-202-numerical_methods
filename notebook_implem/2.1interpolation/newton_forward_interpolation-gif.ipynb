```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#input data

xval = np.linspace(-5,5,18)

#yval = np.e**xval
yval = np.sin(xval) + 0.5 * np.cos(xval) +  0.7 * np.sin(2*xval) 
```

```python
yval
```

```python
#defining constants

h = xval[1] - xval[0]   
n = len(yval)
h
```

```python
#Implementing forward differnce operator

def fdiff(y, n ,i):
    if n == 0:
        return y[i]
    
    return fdiff(y,n-1,i+1) - fdiff(y,n-1,i)
```

```python
#implementing D

def d(x,i):
    ans = 1
    for j in range(0,i):
        ans = ans * (x - xval[j])
    return ans
    
```

```python
d(0,3)
```

```python
#Implementing p for a
def p(i):
    ans = 1
    for j in range(0,i):
        ans = ans * (p - j)
    return ans
```

```python

def coeff(y , order):
    n = len(y)
    c = []
    for i in range(0,order):
        c.append(fdiff(y,i,0) / (math.factorial(i) * h**i))
    return c
```

```python
def pol(coeff , x):
    val = 0
    n = len(coeff)
    for i in range(0,n):
        val = val + coeff[i] * d(x,i)
    return val
```

```python
co = coeff(yval[2:],4)
```

```python
fval = [np.sin(i) + 0.5 * np.cos(i) +  0.7 * np.sin(2*i) for i in np.linspace(-5,5,200)]
pval = [pol(co , i) for i in np.linspace(-5,5,200)]
fig , ax = plt.subplots()
ax.plot(xval,yval , 'ko')
ax.set_xlim([-6, 6])
ax.set_ylim([-10, 10])
ax.plot(np.linspace(-5,5,200), pval , 'blue')
ax.plot(np.linspace(-5,5,200) , fval , 'red')
plt.savefig("image.png")
fig.show()
```

```python
fval = [np.sin(i) + 0.5 * np.cos(i) +  0.7 * np.sin(2*i) for i in np.linspace(-5,5,200)]
xval_i = xval
yval_i = yval
for pi in range(0,len(yval) - 4):
    xval = xval_i[pi:]
    yval = yval_i[pi:]
    co = coeff(yval,4)  
    pval = [pol(co , i) for i in np.linspace(-5,5,200)]
    fig , ax = plt.subplots()
    ax.plot(xval_i,yval_i , 'ko')
    ax.set_xlim([-6, 6])
    ax.set_ylim([-7, 7])
    ax.plot(np.linspace(-5,5,200), pval , 'blue' )
    ax.plot(np.linspace(-5,5,200) , fval , 'red' )
    ax.set_title("polynomial approximation of order 4 of \n $ f(x) = \sin(x) + 0.5\cos(x) + 0.7\sin(2x)$")
    plt.savefig("image{pi}.png".format(pi=pi))
    fig.show()
    
```
