```python
import sympy as sy
from sympy.solvers import solve
from sympy import Eq , lambdify
import scipy as sp
import scipy.optimize
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
x = sy.Symbol('x')
f = x**3 - 2*x + 2
f
```

```python
f_der = sy.diff(f,x)
f_der
```

```python
g = x - f/f_der
sy.simplify(g)
```

```python
phi = sy.simplify(g.subs(x,g))
```

```python
phi1 = sy.simplify(g.subs(x,(g.subs(x,g))))
phi1
```

```python
eq1 = Eq(phi-x,0)
eq1.lhs
```

```python
func = sy.lambdify(x,eq1.lhs)
func
```

```python
func(-0.5846)
```

```python
xval = np.linspace(-2,3,100000)
yval = func(xval)
xaxis = np.zeros(len(xval))
plt.plot(xval,yval)
plt.plot(xval,xaxis)
plt.show()
```

```python
sol = sp.optimize.root(func, 0 )
```

```python
sol.x
```

```python
sol1 = sp.optimize.root(func, 1)

sol1.x
```
